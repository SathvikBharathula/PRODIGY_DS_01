Bar Plot

```python
df = pd.read_csv("/content/worldpopulationdata.csv")

# Filtering only population data
df = df[df['Series Name'] == 'Population, total']

# Selecting required columns
df = df[['Country Name', '2022']]

# Sorting by population (descending)
df_sorted = df.sort_values(by='2022', ascending=False)

# Taking top 10 countries
top10 = df_sorted.head(10)

# Ploting bar plot
plt.figure(figsize=(10,6))
plt.bar(top10['Country Name'], top10['2022'])

# Adding labels and title
plt.title('Top 10 Most Populated Countries (2022)')
plt.xlabel('Country')
plt.ylabel('Population')

# Rotating x-axis labels
plt.xticks(rotation=45)
plt.show()
```

Histogram

```python
df = pd.read_csv('worldpopulationdata.csv')

#Filtering only population data
df = df[df['Series Name'] == 'Population, total']

# Selecting required columns
df['2022'] = df['2022'] / 1e6

# Cleaning data
df = df.dropna()
df['2022'] = pd.to_numeric(df['2022'], errors='coerce')

# Ploting histogram
plt.figure(figsize=(10,6))
plt.hist(df['2022'], bins=20)

# Labels
plt.title('Distribution of Population (2022)')
plt.xlabel('Population (In millions)')
plt.ylabel('Number of Countries')

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Conclusions :

The population distribution across countries is highly uneven, with most countries having relatively small populations while a few countries have extremely large populations.

The histogram shows a right-skewed distribution, indicating that a small number of countries contribute to a large portion of the global population.

The bar chart highlights that countries like India and China dominate in population compared to others, creating a significant gap between the top and remaining countries.

These visualizations help in understanding global population patterns and emphasize the imbalance in population distribution among countries.
